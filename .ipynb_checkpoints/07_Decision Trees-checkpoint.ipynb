# Decision Trees

Als erstes werden wir Decision Trees, auf deutsch Entscheidungsbäume, kennenlernen. Decision Trees sind aufgrund ihrer Einfachheit und Effizienz weit verbreitet. Sie funktionieren ähnlich wie Baumdiagramme und partitionieren einen Datensatz anhand einer Reihe von Entscheidungskriterien. Diese Entscheidungskriterien sind hierarchisch aufgebaut, oftmals in einer baumartigen Struktur. Daher kommt auch der Name Entscheidungsbaum. Aufgrund dieser baumartigen Struktur sind Decision Trees anschaulich und können leicht visualisiert werden. Decision Trees werden ausführlich in der Vorlesung erklärt, daher werden wir hier nicht noch einmal auf alle Details eingehen.

Ein entscheidender Vorteil von Decision Trees ist ihre Fähigkeit, sowohl kategorische als auch numerische Daten zu verarbeiten. Sie können dadurch für Klassifizierungs- als auch für Regressionsprobleme eingesetzt werden. In dieser Sitzung werden wir anhand von zwei Klassifizierungsbeispielen Decision Trees kennenlernen. In der nächsten Sitzung werden wir dann die Regressionsanalyse einführen und diese mit dem Modell der Decision Trees verknüpfen. Die beiden Beispiele sind:

* **Giftige Pilze klassifizieren**: wir haben einen Datensatz mit 22 verschiedenen Eigenschaften von Pilzen und vielen verschiedenen Fällen, die in der Natur gesammelt wurden. Daraus wollen wir ein Modell bauen, welches neue Pilze anhand der Kriterien in giftig und essbar klassifizieren kann.
* **Kreditdaten**: Ein weiterer Datensatz mit Informationen über Personen, die Kredite aufnehmen wollen. Das ist ein klassisches Anwendungsbeispiel, welches in der Praxis häufig verwendet wurde. Eine Bank oder Versicherung benutzt dann ebenfalls solche Modelle, um zu entscheiden, ob eine Person einen Kredit bekommt oder nicht.

#### Supervised Learning

<img src="https://rsample.tidymodels.org/logo.png" alt="rsample" width="100" align="right" /> Im nächsten Schritt wollen wir den Datensatz für das Training eines Modells vorbereiten. Zunächst müssen wir die Daten aufteilen. Bei Modellen, die dem Supervised Learning (Überwachtes Lernen) zuzuordnen sind, benötigen wir mindestens ein "Trainingsset" und ein "Testset". 

- Das **Trainingsset** ist dazu da, dass das Modell lernen/trainieren kann, wie es aufgebaut sein muss, damit es optimale Ergebnisse liefert. Das Modell soll ja später mal aus einer Menge von unabhängigen Variablen (z.B. verschiedene Eigenschaften eines Pilzes) auf eine abhängige Variable schließen (essbar vs.  giftig). Wie der Name schon sagt, trainiert oder "übt" das Modell diese Prognose mit den Daten aus dem Testset ein. Wir "verraten" dem Modell dazu auch direkt die korrekte Lösung. So kann es sich auf die Trainingsdaten anpassen.

- Mit dem **Testset** wollen wir prüfen, wie gut das Modell gelernt hat (Evaluation). Das ist quasi wie eine "Klausur". Das Modell soll mit Ergebnisse vorhersagen und wir vergleichen dann im Anschluss mit den wahren Werten. Wie gut hat das Modell die "Musterlösung" getroffen? 

Theoretisch könnte man ein Modell natürlich auch nur mit einem Trainingsset trainieren und dann damit Prognosen über Daten wagen, bei denen man den wahren Wert der vorherzusagenden Variable gar nicht kennt. Das wäre aber eine ziemlich schlechte Idee: Wir wüssten dann gar nichts über die Qualität der Vorhersagen, die das Modell macht! Es kann sein, dass das Modell einfach sehr gut auf unsere Trainingsdaten angepasst ist, die neuen Daten aber davon abweichen. Der Test des Modells und Aussagen über die Güte der Vorhersage (Performance), sind deshalb immer Bestandteil der Modellierung. Supervised Learning und die Probleme die beim Over- bzw. Underfitting entstehen, werden ausführlicher in der Vorlesung erklärt. 

Um die vorliegenden Daten in Trainings- und Testset aufzuteilen, beinhaltet `tidymodels` das Paket [`rsample`](https://rsample.tidymodels.org/). In `rsample` können wir mit der Funktion `initial_split()` den ursprünglichen Datensatz aufteilen und dann über `training()` und `testing()` die Trainings- und Testdaten festschreiben. Ein typischer Anteil für die Aufteilung sind 75% Trainingsdaten und 25% Testdaten - das können wir mithilfe des Arguments `prop = 0.75` übergeben.

> Die Auswahl von `initial_split()` ist zufällig. Um  **reproduzierbare** Ergebnisse zu bekommen, setzen wir mithilfe von `set.seed(123)` den Random Number Generator fest, sodass die Aufteilung der Daten bei jedem Durchlauf des R-Skriptes und auf jedem Rechner gleich bleibt. Ansonsten würden wir immer leicht unterschiedliche Ergebnisse erhalten.

```r
# set a seed for reproducibility
set.seed(123)

# Daten in Trainings- und Testdaten aufteilen
data_split <- mushroom %>%
    initial_split(prop = 0.75) # 75% der Daten werden zu Trainingsdaten

# Trainingsdaten festschreiben
train_data <- data_split %>%
    training()

# Testdaten festschreiben
test_data <- data_split %>%
    testing()
```

### Modell trainieren

<img src="https://parsnip.tidymodels.org/logo.png" alt="parsnip" width="100" align="right"/>  Wir trainieren das Modell darauf, giftige von essbaren Pilzen zu unterscheiden. Diese Information ist in der Variable `type` gespeichert! Zuallererst müssen wir aber ein Decision Tree Modell erstellen. Dafür können wir die Funktion [`decision_tree()`](https://parsnip.tidymodels.org/reference/decision_tree.html) aus dem Paket `parsnip` nutzen. Aus der offiziellen Dokumentation können wir entnehmen:

> `decision_tree()` defines a model as a set of if/then statements that creates a tree-based structure. This function can fit classification, regression, and censored regression models.

Zunächst müssen wir die **Eigenschaften des Modells definieren**. Als computational engine wollen wir `"rpart"` nutzen. `rpart` ist ebenfalls ein weit verbreitetes R-Paket für *recursive partitioning models*, also auch für Decision Trees. Da wir mit unseren Pilzen ein Klassifizierungsproblem vorliegen haben, setzen wir als Mode noch `"classification"`. Das definierte Modell (klassifizierender Decision Tree mit der Engine `rpart`) speichern wir anschließend in dem Objekt `tree_spec`.

```r
# Modell definieren
tree_spec <- decision_tree() %>% # definiere das Modell als Decision Tree
    set_engine("rpart") %>% # set rpart as enginge
    set_mode("classification") # we want to solve a classification problem
```

Jetzt haben wir die Eigenschaften des Modells festgelegt, aber das Modell hat noch keine Daten gesehen und konnte noch nichts trainieren. Das ist der nächste Schritt: Wir das **fitten das Modell**. D.h., wir lassen das Modell anhand unserer Trainingsdaten lernen. Der Decision Tree lernt also, die Variable `type` (also die Essbarkeit/Giftigkeit der Pilze) zu klassifizieren, und zwar anhand einer Reihe von anderen Variablen, die ebenfalls im Datensatz sind. Dazu benutzen wir nur den Teil-Datensatz, der in den Trainingsdaten enthalten ist. Für das Fitting nutzen wir die Funktion `fit()`, ebenfalls aus `parsnip`.

`fit()` können wir eine Formel übergeben, und zwar `type ~ .`. Damit sagen wir `fit()`, dass `type` unsere zu vorhersagende Variable ist. Mit `.` sagen wir, dass alle anderen Variablen als Prädiktoren (predictors) genutzt werden sollen. Außerdem müssen wir dem Modell mit `data = train_data` vorgeben, dass auf den Trainingsdaten trainiert werden soll.

> Der Operator `~` wird in R benutzt, um die Beziehung zwischen abhängigen und unabhängigen Variablen in einem statistischen Modell zu definieren. Auf der linken Seite steht die abhängige Variable, auf der rechten Seiten die sogenannten unabhängigen Variablen.

```r
# Das Modell auf die Trainingsdaten fitten
tree_fit <- tree_spec %>%
    fit(
        type ~ ., # type soll durch alle anderen Variablen vorhergesagt werden
        data = train_data # Wir benutzen train_data zum Training
    )
```

Mithilfe von `tree_fit$fit` können wir uns die Regeln ausgeben lassen, die das Modell ermittelt hat.

#### `rpart.plot()` 

Wir können die Regeln aus `tree_fit$fit` auch darstellen! Leider gibt es dafür keine Funktion innerhalb von `tidymodels`, es gibt aber die Funktion `rpart.plot()` aus dem gleichnamigen Paket.

```r
# Decision Tree visualisieren
tree_fit$fit %>%
    rpart.plot(
        type = 4, # Typ des Plot - Wert 4 erzeugt gezeichneten Baum mit abgerundeten Ecken und variabler Breite der Zweige
        extra = 1, # Zeige die Anzahl der Fälle in jedem Knoten und den Prozentsatz der Beobachtungen im Knoten im Vergleich zur Gesamtzahl der Beobachtungen
        cex = 1.5, # Größe des Textes / der Labels
        roundint = FALSE # Zeige die Aufteilungspunkte als Fließkommazahlen anstatt als ganze Zahlen
    )
```

Aus der Textdarstellung über `tree_fit$fit` können wir entnehmen, dass unser Modell nur 5 Regeln aufstellt. Auch in der Visualisierung lässt sich erkennen, dass das Modell nicht besonders kompliziert wird.

#### Neue Daten vorhersagen

Unser Modell funktioniert jetzt bereits und kann tatsächlich bereits Prognosen treffen. Wir laufen also in den Wald und finden einen unbekannten Pilz. Wir schauen uns den Pilz grob an und erfassen die Variablen, die für das Modell relevant sind in einem Datensatz - sind ja nur zwei, `odor` und `spore_print_color`. Beim Rest können wir einfach `NA` eintragen, ist ja nicht wichtig. Diesen Datensatz `new_pilz` übergeben wir unserem Decision Tree Modell `tree_fit` der Funktion `predict()` aus `parsnip`.

```r
new_pilz <- tibble(
    cap_shape = NA,
    cap_surface = NA,
    cap_color = NA,
    bruises = NA,
    odor = "almond",
    gill_attachment = NA,
    gill_spacing = NA,
    gill_size = NA,
    gill_color = NA,
    stalk_shape = NA,
    stalk_root = NA, 
    stalk_surface_above_ring = NA,
    stalk_surface_below_ring = NA,
    stalk_color_above_ring = NA,
    stalk_color_below_ring = NA, 
    veil_type = NA,
    veil_color = NA, 
    ring_number = NA,
    ring_type = NA,
    spore_print_color = "brown",
    population = NA,
    habitat = NA
    ) %>%
    mutate_all(as.factor)

tree_fit %>% # take the model
    predict(new_pilz) %>% # predict our newly defined mushroom
    pull(.pred_class) # pull the class out of the prediction
```

Voilà, laut Modell ist unser Pilz essbar! Bevor wir den Pilz in der Pfanne zubereiten, sollten wir uns allerdings erst einmal fragen, wie gut eigentlich die Vorhersagen unseres Modells sind.

## Decision Trees

### Modell evaluieren

<img src="https://parsnip.tidymodels.org/logo.png" alt="parsnip" width="100" align="right"/> Um die Performance eines statistischen Modells zu evaluieren, gibt es in R verschiedene Pakete, z.B. `caret` oder `mlr`. Wir benutzen an dieser Stelle das Paket `parsnip`, welches ebenfalls Teil von `tidymodels` ist. Wir erinnern uns: einige Schritte vorher hatten wir die vorhandenen Daten in Trainings- und Testdaten aufgeteilt. Wir haben mittlerweile das Modell mithilfe der Trainingsdaten trainiert und wollen als nächstes gucken, ob wir mit unserem Modell auch gute Vorhersagen treffen können. 

Dafür wenden wir zunächst wieder `predict()` auf `tree_fit` an, nehmen aber diesmal die gesamten Testdaten `test_data` als Datengrundlage. Die vorhergesagten Ergebnisse werden in die neue Spalte `predicted` geschrieben und das ganze wird in einem neuen Tibble `results` gespeichert.


```r
# directly compare the test data with the predicted data?
results <- test_data %>% # auf den Testdatensatz anwenden
  mutate(
      predicted = predict( # eine Spalte predictions hinzufügen, die die Vorhersagen beinhaltet
          tree_fit, # wir wollen das Modell tree_fit nehmen
          test_data, # und damit den Testdatensatz vorhersagen
          type = "class" # es handelt sich um ein Klassifizierungsproblem
      )$.pred_class # aus unserer Vorhersage predictions interessiert uns die Variable .pred_class
  )
```

#### Confusion Matrix

<img src="https://yardstick.tidymodels.org/logo.png" alt="parsnip" width="100" align="right"/> Anschließend wollen wir eine sogenannte **Confusion Matrix** erstellen. Vielleicht ist euch diese bereits in der Vorlesung begegnet. Sie bietet uns einen schnellen Überblick darüber, inwiefern unser Modell die richtigen oder falschen Labels vorhersagt. Für die Evaluation von Modellen ist in `tidymodels` das Paket `yardstick` beinhaltet. Dieses gibt uns mit der Funktion `conf_mat()` direkt eine Möglichkeit, eine Confusion Matrix zu erstellen.

Dafür müssen wir zunächst unser Modell auf den Testdatensatz `test_data` anwenden. Dafür benutzen wir wie oben die Funktion `predict()` und fügen die Ergebnisse als neue Spalte `predicted` direkt hinzu. Das Ergebnis speichern wir in einem neuen Datensatz `results`. Wenn ihr möchtet, könnt ihr euch `results` einmal mit `glimpse()` ausgeben lassen, dann seht ihr genauer, was passiert ist.

Anschließend übergeben wir die Ergebnisse an `conf_mat()`. Die Evaluation funktioniert dann folgendermaßen: Im Testdatensatz sind die tatsächlichen Kategorien bereits in `type` gelabelt. Das ist quasi die "Wahrheit", oder `truth`. Die Vorhersage `estimate` erhalten wir über die vorhergesagte Spalte `predicted`. Die tatsächlichen und die vorhergesagten Werte können wir jetzt in der Confusion Matrix miteinander vergleichen.

```r
# create a confusion matrix
results %>%
    conf_mat(truth = type, estimate = predicted)
```

7. Wie interpretiert ihr die Tabelle?

Wir haben auf der linken Seite die Vorhersagen `Prediction` und oben die tatsächlichen Werte `Truth`. Leider scheint unser Modell nicht perfekt zu sein und Fehler zu machen und zwar in eine ganz bestimmte Richtung. Sind diese Fehler akzeptabel?

#### Metrics

Über die Confusion Matrix können wir bereits die wichtigsten Aspekte herauslesen. Bei anderen Datensätzen ist es aber nicht so verkehrt, allgemeine statistische Aussagen über die Performance des Modells treffen zu können. Dafür gibt es verschiedene Metriken, zum Beispiel die `accuracy`. Das ist im Prinzip die prozentuelle Genauigkeit unseres Modells. 

8. Wie würdet ihr die `accuracy` unseres Modells händisch mithilfe der Confusion Matrix berechnen?

In `yardstick` sind glücklicherweise Funktionen vorhanden, mit den wir verschiedene Metriken komfortabel berechnen können. Wir können zunächst einen Satz aus verschiedenen Metriken erstellen, um das Modell mit unetrschiedlichen Perspektiven beurteilen zu können. Uns interessiert der Einfachheit halber erstmal nur die `accuracy`. Das Erstellen des Satzes passiert über `metric_set(accuracy)`. Diesen Satz speichern wir im Objekt `metrics`.   

Anschließend fügen wir dem Testdatensatz wieder eine Spalte mit den Vorhersagen `predictions` hinzu, so wie wir es weiter oben schon gemacht haben. Dann wenden wir den Satz aus Metriken auf diesen Datensatz an. `yardstick` erkennt automatisch, dass es den von uns definierten Satz aus Metriken verwendet, wir müssen lediglich wieder die `truth` (tatsächliche Werte) und den `estimate` (Vorhersagen) angeben.

```r
metrics <- metric_set(accuracy) # Satz aus Metriken definieren

model_performance <- test_data %>%
    mutate(predictions = predict(tree_fit, test_data, type = "class")$.pred_class) %>% # Vorhersagen zum Testdatensatz hinzufügen
    metrics(truth = type, estimate = predictions) %>% # Satz aus Metriken darauf anwenden
    print() 
```

`yardstick` erkennt, dass es sich bei unserer vorherzusagenden Variable um eine "binäre" Variable (also eine Variable mit nur zwei Ausprägungen) handelt und berechnet uns eine `accuracy` von 0.993. Das entspricht auch der `accuracy`, die wir händisch berechnen würden.

$accuracy = 1 - 15 / (1068 + 15 + 948) \thickapprox 0.993$ 

#### Variable Importance Plot

Eine weitere interessante Darstellung zur Evaluierung eines Decision Trees ist der sogenannte *Variable Importance Plot* (VIP). Dieser stellt die relative Bedeutung der Variablen innerhalb eines statistischen Modells dar. Dadurch können wir sehen, welche Variablen den **größten Einfluss** auf die Vorhersageleistung des Modells haben! Je nach Modell gibt es verschiedene Berechnungsweisen zur Bedeutung der Variablen. Auf die konkrete Berechnung werden wir hier aber nicht weiter eingehen.

Eine ausführliche mathematische Beschreibung zur Berechnungsweise eines Variable Importance Plots findest du z.B. [hier](https://koalaverse.github.io/vip/articles/vip.html).

Ein VIP ist nützlich, um die relevantesten Variablen in einem Modell zu identifizieren und einen Einblick in den Zusammenhang zwischen Variablen und Zielgröße (in unserem Pilzbeispiel dann die Variable `type`) verschaffen. Gleichzeitig ist ein VIP auch mit Vorsicht zu genießen. So kann es häufiger vorkommen, dass es starke Korrelationen zwischen einzelnen Parametern gibt - also beispielsweise ein Parameter stark von einem anderen abhängt. Ein Beispiel: wenn ein Pilz immer an der Nordseite eines Baumes wächst, dann ist die Wahrscheinlichkeit sehr hoch, dass dieser Pilz wenig Sonne abbekommt. In einem VIP muss man dann differenzieren, welche Variable jetzt der eigentliche Treiber des Modells ist. Der Standort oder die Sonnendauer pro Tag?

In R können wir VIP mithilfe des Package `vip` erstellen. Im Folgenden wird ein VIP für unser weiter oben trainiertes Modell erzeugt. Dafür müssen wir zunächst die Variable Importance für das Modell mithilfe der Funktion `vip::vi()` berechnen:

```r
# create a variable importance dataframe using vip::vi()
vis = vi(
    tree_fit, # we want to determine the variable importance of our model tree_fit
    scale = TRUE) %>% # scale the most important variable to 100
    print()
```

Hier können wir bereits sehen, dass die wichtigsten beiden Variablen der Geruch `odor` und die Farbe der Sporen `spore_print_color` sind. Wir können die Tabelle auch der Anschaulichkeit halber nochmal visualisieren:

```r
# plot the results using vip::vip()
vis %>%
    vip(num_features = 6) + # plot only the six most important features
    theme_classic(base_size = 25)
```

9. Vergleicht den Variable Importance Plot mit den Regeln, die wir weiter oben mit `tree_fit$fit` ausgegeben haben. Stimmen die Ergebnisse des Variable Importance Plots mit den Regeln des Entscheidungsbaumes überein? 

### Modell verbessern

Im Kontext des Machine Learning gibt es viele Techniken, mit denen ein Modell verbessert werden kann. An dieser Stelle wollen wir allerdings erstmal nur eine Methode vorstellen, und zwar die der **Kostenfunktion**. Die Idee dabei ist einfach, dass bestimmte Ergebnisse des Modells unerwünscht sind, und sozusagen große "Kosten" verursachen. Beim Beispiel der Pilze ist das einfach: wenn unser Modell `edible` vorhersagt, der Pilz aber `poisonous` ist, dann kann das sehr negative Konsequenzen haben. Diesen Fall wollen wir also unbedingt vermeiden! Auf abstrakter Ebene heißt das, dass wir dem Modell einen Anreiz geben, dass dieser Fall nicht eintritt! Jedes Ergebnis verursacht jetzt zusätzlich noch Kosten, die das Modell versucht, zu minimieren. Wenn wir also einen großen Kosten-Faktor für den oben genannten Fall einführen, dann wird das Modell lernen, diese Fälle zu vermeiden!

Glücklicherweise kann man der `rpart`-Engine von `decision_tree()` direkt eine Kostenfunktion übergeben! Dafür definieren wir zunächst eine `cost_matrix`. Diese entspricht im Prinzip der Confusion Matrix von weiter oben.

|   | *Truth*: Edible | *Truth*: Poisonous |
|-------------|--------|-----------|
| ***Prediction***: Edible      | Fall A | Fall B    |
| ***Prediction***: Poisonous   | Fall C | Fall D    |

Fall A und Fall D sind optimal für uns, weil das Modell dann eine richtige Vorhersage gemacht hat. Der Kostenfaktor dieser beiden Fälle ist 0 - wir wollen das Modell ja nicht dafür bestrafen, dass es eine richtige Vorhersage macht! Fall C ist zwar eine falsche Vorhersage, für uns aber erstmal nicht so schlimm. Lieber einen Pilz als giftig einsortieren, der eigentlich essbar war, als andersrum! Daher belassen wir den Kostenfaktor bei 1 (das entspricht dem Standardwert). Fall B ist richtig gefährlich, wir müssen unbedingt vermeiden, dass Pilze als essbar kategorisiert werden, die eigentlich giftig sind. Daher wählen wir für Fall B einen großen Kostenfaktor, z.B. 10!

```r
cost_matrix <- matrix(c(0, 10, 1, 0), nrow = 2)
```

Diese `cost_matrix` können wir jetzt innerhalb der Funktion `set_engine()` direkt an das Modell übergeben und wieder das Modell auf den Testdatensatz trainieren.

```r
tree_fit_improved <- decision_tree() %>% # wir erstellen ein neues Modell
    set_engine("rpart", parms = list(loss = cost_matrix)) %>% # an dieser Stelle übergeben wir die Kostenmatrix!
    set_mode("classification") %>%
    fit(type ~ ., data = train_data)
```

Wenn wir uns jetzt die Regeln unseres verbesserten Modells anschauen, merken wir, dass einige hinzugekommen sind:

```r
tree_fit_improved$fit
```

Das liegt auf der Hand, denn das Modell muss sich jetzt zusätzliche Regeln ausdenken, um Fall B um alle Fälle zu vermeiden (10 ist ein relativ hoher Kostenfaktor).

Wie sieht es mit der Confusion Matrix aus?

```r
# directly compare the test data with the predicted data?
results_improved <- test_data %>%
  mutate(predicted = predict(tree_fit_improved, test_data, type = "class")$.pred_class)

# create a confusion matrix
results_improved %>%
    conf_mat(truth = type, estimate = predicted)
```

Voilà. Wir haben Fall B komplett ausgeschlossen. Vermutlich werden bei einem größeren Datensatz jetzt einige Pilze als `poisonous` einsortiert, dafür vermeiden wir aber Fall B! Wir haben das Modell also mithilfe einer **Kostenfunktion** verbessert.

Im Laufe von Block 2 werden wir noch einige weitere Möglichkeiten kennenlernen, wie Maching Learning Modelle verbessert werden können. Die Kostenfunktion wird uns immer wieder begegnen, denn sie spielt eine große Rolle bei neuronalen Netzen.
